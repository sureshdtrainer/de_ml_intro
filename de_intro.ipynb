```python
import pandas as pd
```

```python
#Data Set Overview

#read the CSV 
df_netflix = pd.read_csv('netflix_titles.csv')
#df_netflix.tail()
#df_netflix.head()
#df_netflix.dtypes
#df_netflix.info()
#df_netflix.describe()
#df_netflix.values
```

```python
# identify missing data

#number of rows missing in each column
#df_netflix.isnull().sum()

#to get top most column with null values
df_netflix.isnull().sum().sort_values(ascending=False)
#list of column
#df_netflix.columns
```

```python
#the mean value for director column with null
df_netflix['director'].isnull().mean()

#get the percentage of each column in dataset with null values
for column in df_netflix.columns:
    percentage_null= df_netflix[column].isnull().mean()*100
    print(column + ': ' +str(percentage_null))
```

```python
# Handle the director column as it has most number of null values
#How to deal with missing data
#1. Drop the columns with bad data
df_netflix.drop('director', axis=1)
```

```python
# 2. Drop the complete rows with null values
no_director=df_netflix[df_netflix['director'].isnull()].index
df_netflix.drop(no_director, axis=0)
```

```python
#to acheive the same behaviour using ~ is null (Not null)
#df_netflix[~(df_netflix['director'].isnull())]
# dropna
df_netflix.dropna(subset=['director'])
#check to get the count of nulls in all columns
df_netflix.dropna(subset=['director']).isnull().sum()
```

```python
#3. fillna() to replace NaN by the mean, median(numeric value) or mode
#1. use the most common value using mode function
df_netflix[df_netflix['rating'].isnull()] #10 rows with rating as NaN

df_netflix['rating'].mode() #most common value
mode=''.join(df_netflix['rating'].mode())
#type(mode)
df_netflix['rating'].fillna(mode)

#check if null exists
df_netflix['rating'].fillna(mode).isnull().sum()
```

```python
#2. to use back fill (bfill) or front fill (ffill)
df_netflix['director']
```

```python
#df_netflix['director'].ffill()
df_netflix['director'].bfill()
df_netflix.ffill().isnull().sum()
```

```python
#Extracting the data split or extract
#Extract the duration int value in minutes from Movies only from the below data set
df_movie=df_netflix[df_netflix['type']=='Movie'] #filter only movies rows
df_movie['minute']=df_movie['duration'].str.split(expand=True)[0].astype(int) #extracting the minutes value from the column
df_movie
```

```python
df_movie['year_added']=df_movie['date_added'].str.split(',',expand=True)[1].astype(str)
df_movie
```
